```python
import pandas as pd
```

```python
df = pd.read_excel('vocab.xlsx')
```

```python
df
```

```python
df.dropna(how='all')
```

```python
df = df.dropna(how='all')
```

```python
df = df[df['序号']!='序号']
```

```python
df
```

```python
df.columns
```

```python
df[['序号.1', '英文.1', '中文.1']].rename(columns={'序号.1':'序号', '英文.1':'英文', '中文.1':'中文'}).dropna()
```

```python
df = pd.concat([df[['序号', '英文', '中文']], df[['序号.1', '英文.1', '中文.1']].rename(columns={'序号.1':'序号', '英文.1':'英文', '中文.1':'中文'}).dropna()], axis=0)
```

```python
df
```

```python
df = df.set_index('序号')
```

```python
df = df.reset_index(drop=True)
```
